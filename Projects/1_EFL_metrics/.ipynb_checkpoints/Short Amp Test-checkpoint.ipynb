```python
alld=[]     # this is all the data
allm=[]     # these are all the markers. Markers are given relative to the beginning of the data chunk you just got
            # so you have to do a calculation to give the markers relative to the start of the recording.


# For the ELF, you should probably increase the amount of time. Right now it runs until the end
# but I have to figure out a way to stop it remotely. You can always press CTRL-C. But not sure that works.

t0=time.time()
ctime=t0
# while ctime - t0 < 500:
while True:    
    ctime=time.time()
    time.sleep(2.0);
    print('-')
    
    if (ctime-t0) % 10 == 0:
        print('%d seconds passed' % 10)
    
    # this is where you get the data
    data, marker = amp.get_data()

    # append it to the big matrix
    alld.append(data)
    for m in marker:
        allm.append(m)
    
    # print('---')
    # print(data.shape)
    # print(marker)
    
    # figure out a way to stop this without KeyboardInterrupt
```

### json = header

```python
meta=json.load(fhj)
print(meta)
```

```python
meta['Channels']
```

### binary = raw data

```python
import numpy as np

f = open("t18.eeg", "r")
m = np.fromfile(f, dtype=np.float32)
```

```python
len(m)/64
```

```python
m=m.reshape(round(len(m)/64),64).transpose()
```

```python
import mne
```

```python
# Generate some random data

# Initialize an info structure
info = mne.create_info(
    ch_names=meta['Channels'],
    ch_types = ['eeg' for i in range(64)],
    sfreq=meta['Sampling Frequency']
)

custom_raw = mne.io.RawArray(m, info)
print(custom_raw)
```

### markers

```python
with open('t18.marker') as file:
    content = file.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content] 
print(content)

import re

# create the marker matrix
ev_arr=[]
for i, item in enumerate(content):
    out=re.split("[\s|S|T]+", item)
    if 'Sync Off' in item:
        sample = int(float(out[0])/1000*meta['Sampling Frequency'])
        code = 250
    else:
        sample = int(float(out[0])/1000*meta['Sampling Frequency'])
        code = int(out[1])

    # print(out)
    # print(sample)
    ev_arr.append([sample, 0, code])
```

```python
l=[x[0] for x in ev_arr]
np.diff(l)
```

```python
# create marker channel for MNE python:
if ev_arr:
    info = mne.create_info(['STI'], custom_raw.info['sfreq'], ['stim'])
    stim_data = np.zeros((1, len(custom_raw.times)))
    stim_raw = mne.io.RawArray(stim_data, info)
    custom_raw.add_channels([stim_raw], force_update_info=True)

    # create the marker matrix:
    custom_raw.add_events(ev_arr)
```
